# Preprocessing



### Read data

```python
with open('../data/pg1597.txt') as f:
    lines = f.readlines()
print(lines[:6])
```

### Tokenization

```python
# import nltk
# nltk.download('punkt')
from nltk.tokenize import word_tokenize
data = "".join(lines)
data = word_tokenize(data)
print(data[:51])
```

### Lower case

```python
words = " ".join(data)
lower_w = words.lower()
lower_w[:50]
```

### Stop words removal

```python
# import nltk
# nltk.download('stopwords')
from nltk.corpus import stopwords 
  
stop_words = set(stopwords.words('english')) 
word_tokens = lower_w.split(" ")
  
filtered_sentence = [w for w in word_tokens if not w in stop_words] 
# filtered_sentence[0] = filtered_sentence[0][1:]
print(filtered_sentence[:51])
```

### Stemming

```python
import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer()

output = []
for word in filtered_sentence:
    output.append((ps.stem(word)))
```

```python
final_out = " ".join(output)
```

```python
print(final_out[:500])
```

### Vocabulary

```python
def get_vocab(data):
    data = data.replace('\n', ' ').split(' ')
    words = list(set(data))
    vocabulary = {word:index for index, word in enumerate(words)}
    vocab_size = len(vocabulary)
    
    return vocabulary, vocab_size
```

```python
vocabulary, vocab_size = get_vocab(final_out)
vocabulary
```
